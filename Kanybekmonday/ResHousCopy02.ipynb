```python
mm = folium.Map(location=[29, -95],zoom_start=3)
from folium.plugins import MarkerCluster

mcg = folium.plugins.MarkerCluster(control=False)
mm.add_child(mcg)

Mex = folium.plugins.FeatureGroupSubGroup(mcg, 'Mex') 
mm.add_child(Mex)

FF = folium.plugins.FeatureGroupSubGroup(mcg, 'FF') 
mm.add_child(FF)

Pizza = folium.plugins.FeatureGroupSubGroup(mcg, 'Pizza')
mm.add_child(Pizza)

Sandwiches = folium.plugins.FeatureGroupSubGroup(mcg, 'Sandwiches') 
mm.add_child(Sandwiches)

Burgers = folium.plugins.FeatureGroupSubGroup(mcg, 'Burgers')
mm.add_child(Burgers)

# Mex = MarkerCluster().add_to(mm)
for i in range(0, len(dfMex)):
    folium.CircleMarker([dfMex.iloc[i]['latitude'], dfMex.iloc[i]['longitude']],radius=dfMex.iloc[i]['rating']*2,
                 popup=str('Rating:  ')+str(dfMex.iloc[i]['rating'])+"<br>"+ "<b>"+str(dfMex.iloc[i]['name'])+ "</b>"+"<br>"+
                  "<br>"+str('address: ')+str(dfMex.iloc[i]['display_address']), color='blue', fill=True, fill_color='blue').add_to(Mex)

                                                            
# FF = MarkerCluster().add_to(mm)
for i in range(0, len(dfFF)):
    folium.CircleMarker([dfFF.iloc[i]['latitude'], dfFF.iloc[i]['longitude']],radius=dfFF.iloc[i]['rating']*2, 
                 popup=str('Rating:  ')+str(dfFF.iloc[i]['rating'])+"<br>"+ "<b>"+str(dfFF.iloc[i]['name'])+ "</b>"+"<br>"+
                  "<br>"+str('address: ')+str(dfFF.iloc[i]['display_address']), color='blue', fill=True, fill_color='blue').add_to(FF)

# Pizza = MarkerCluster().add_to(mm)
for i in range(0, len(dfPiz)):
    folium.CircleMarker([dfPiz.iloc[i]['latitude'], dfPiz.iloc[i]['longitude']],radius=dfPiz.iloc[i]['rating']*2, 
                 popup=str('Rating:  ')+str(dfPiz.iloc[i]['rating'])+"<br>"+ "<b>"+str(dfPiz.iloc[i]['name'])+ "</b>"+"<br>"+
                  "<br>"+str('address: ')+str(dfPiz.iloc[i]['display_address']), color='blue', fill=True, fill_color='blue').add_to(Pizza)
    
# Sandwiches = MarkerCluster().add_to(mm)
for i in range(0, len(dfSand)):
    folium.CircleMarker([dfSand.iloc[i]['latitude'], dfSand.iloc[i]['longitude']],radius=dfSand.iloc[i]['rating']*2, 
                 popup=str('Rating:  ')+str(dfSand.iloc[i]['rating'])+"<br>"+ "<b>"+str(dfSand.iloc[i]['name'])+ "</b>"+"<br>"+
                  "<br>"+str('address: ')+str(dfSand.iloc[i]['display_address']), color='blue', fill=True, fill_color='blue').add_to(Sandwiches)
    
# Burgers = MarkerCluster().add_to(mm)
for i in range(0, len(dfBurger)):
    folium.CircleMarker([dfBurger.iloc[i]['latitude'], dfBurger.iloc[i]['longitude']],radius=dfBurger.iloc[i]['rating']*2, 
                 popup=str('Rating:  ')+str(dfBurger.iloc[i]['rating'])+"<br>"+ "<b>"+str(dfBurger.iloc[i]['name'])+ "</b>"+"<br>"+
                  "<br>"+str('address: ')+str(dfBurger.iloc[i]['display_address']), color='blue', fill=True, fill_color='blue').add_to(Burgers)
                  
folium.LayerControl().add_to(mm)



mm.save('restmap.html')
mm.save('index.html')
```
